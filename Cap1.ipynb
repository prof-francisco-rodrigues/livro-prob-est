```python
# Pág. 10
import numpy as np
import matplotlib.pyplot as plt
import random

vf = [] # lista que armazena a frequência de ocorrências
vsim = [] # armazena o número de simulações
Nmax = 1000 # número máximo de simulações
p = 0.5 # probabilidade teórica
moeda = ['cara', 'coroa']
for ns in np.arange(1,Nmax+1): # simula de 1 até Nmax
    nc = 0 # número de caras
    for i in range(0,ns): # simula nsim lançamentos
        saida = random.choice(moeda) # saída da moeda
        if(saida == 'cara'):
            nc = nc + 1
    vf.append(nc/ns) # armazena a frequência
    vsim.append(ns) # armazena o número de experimentos
# mostra os resultados em um gráfico
plt.figure(figsize=(8,6))
plt.plot(vsim, vf, linestyle='-', color="gray", 
         linewidth=2,label = 'Valor simulado')
plt.axhline(y=p, color='black', linestyle='--', 
            label = 'Valor teórico')
plt.ylabel("Fração de caras", fontsize=20)
plt.xlabel("Número de experimentos", fontsize=20)
plt.xlim([0.0, Nmax])
plt.ylim([0.0, 1.0])
plt.legend(fontsize = 15)
plt.show(True)
```

```python
# Pág. 12
import random
# B: bola branca, P: bola preta
# consideramos 3 mil bolas brancas e 2 mil pretas
urna = ['B']*3000 + ['P']*2000 # cria a lista urna
nsim = 100 # número de retiradas
fA = 0 # frequência de bolas pretas
for i in range(0,nsim):
    bola = random.choice(urna) # sorteia uma bola
    if(bola == 'P'): # se a bola for preta
        fA = fA + 1
fA = fA/nsim
print('Frequência de bolas pretas:', fA)
```

```python
# Pág. 28
import random
# define a semente do gerador de números aleatórios
random.seed(10) 
muda_porta = True # decide se muda a porta
if(muda_porta == True):
    print('Opção: Muda a porta')
else:
    print('Opção: Não muda a porta')
print('*************')
ns = 10 # número de simulações
vitoria = 0 # número de vezes que ganha o carro
derrota = 0 # número de derrotas
for s in range(0,ns):
    print('Round:', s+1)
    portas = [1,2,3] 
    # coloca o carro em uma porta
    porta_car = random.choice(portas)
    # seleciona uma porta
    porta_sel = random.choice(portas)
    print('Porta selecionada:', porta_sel)
    print('Porta com o carro:', porta_car)
    # se o carro não está na porta selecionada
    if(porta_sel != porta_car):
        portas.remove(porta_sel)
        portas.remove(porta_car)
        porta_open = portas.pop()
        print('Porta aberta:', porta_open)
        # se mudar, vence
        if(muda_porta == True):
            vitoria = vitoria + 1
            print('Muda a porta: vitória')
        # se não mudar, perde o carro
        else:
            derrota = derrota + 1
            print('Não muda a porta: derrota')
    else:
        # se a porta selecionada contém o carro
        portas.remove(porta_sel)
        porta_open = random.choice(portas)
        print('Porta aberta:', porta_open)
        # se mudar, perde o carro
        if(muda_porta == True):
            derrota = derrota + 1
            print('Muda a porta: derrota')
        # se não mudar, vence
        else:
            vitoria = vitoria + 1
            print('Não muda a porta: vitória')
    print('*************')
print('Fração de vitórias:', vitoria/ns)
print('Fração de derrotas:', derrota/ns)
```

```python
# Pág. 35
import numpy as np

# número de pessoas
n = 30 
# número de simulações
ns = 1000
coincidencias = 0
for _ in range(ns):
    # sorteia aniversários entre 1 e 365
    aniversarios = np.random.randint(1, 366, size=n)
    # verifica se há pelo menos uma coincidência
    if len(aniversarios) != len(np.unique(aniversarios)):
        coincidencias += 1

# probabilidade estimada
prob_estimada = coincidencias / ns
print('Probabilidade estimada:', prob_estimada)
```
