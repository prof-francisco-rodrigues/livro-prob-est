```python
# Pág. 47
import random
# semente do gerar de números aleatórios
random.seed(2007) 
aposta = 10
Ns = 100 # número de jogos
p = 0
for n in range(Ns):
    G = 0 # ganho do jogador
    j = random.randint(1, 6) # valor do jogador
    b = random.randint(1, 6) # valor do oponente
    if(j > b): # se o valor do jogador é maior
        G = 3*(j-b)
    if(G > aposta): # se houver lucro
        p = p + 1
p = p/Ns # probabilidade de ter lucro
print('Aposta:', aposta, 'reais.')
print("Probabilidade de lucro:", p)
```

```python
# Pág. 59
from random import randint
import numpy as np
import matplotlib.pyplot as plt

# define a semente do gerador de números aleatórios
np.random.seed(42)

medias = [] # armazena o valor esperado
S = range(1,1000) # número de simulações
for s in S: # simula o lançamento
    saidas_dado = []
    for n in range(0,s):# para s simulações
        face = randint(1, 6) # simula um lançamento
        # guarda o valor na face do dado
        saidas_dado.append(face)
    # calcula o valor esperado
    val_esp = 0
    for x in saidas_dado:
        val_esp = val_esp + x
    val_esp = val_esp/s
    # armazena o valor esperado
    medias.append(val_esp) 
# mostra os resultados
plt.figure(figsize=(8,6))
plt.plot(S, medias, linestyle='-', color="gray", 
         linewidth=2,label = 'Valor simulado')
plt.axhline(y=3.5, color='black', 
            linestyle='--', label = 'Valor teórico')
plt.xlabel("Número de experimentos", fontsize=20)
plt.ylabel("Valor esperado", fontsize=20)
plt.legend(fontsize = 12)
plt.show(True)
```

```python
# Pág. 60
import numpy as np

Ns = 1000 # número de simulações
p = 0.8 # probabilidade de vender um pão
lucro = 1 # lucro por pão
prejuizo = 0.5 # prejuízo por pão
ganho = 0 # lucro médio
for s in range(0,Ns):
    # se vende um pão
    if np.random.uniform() < p:
        ganho = ganho + lucro
    # caso contrário
    else:
        ganho = ganho - prejuizo
print('Lucro médio: ', ganho/Ns)
```

```python
# Pág. 80
import random
random.seed(17) 

# cria a urna com os números de 1 a 10
urna = list(range(1, 11))
ns = 10000 # número de simulações
soma = 12
k = 0 # contador
for i in range(ns):
    # sorteia dois elementos distintos da urna 
    x1, x2 = random.sample(urna, 2) # (sem reposição)
    # verifica se a soma dos dois valores é igual a 12
    if x1 + x2 == soma:
        k = k + 1 # incrementa o contador
print('Probabilidade estimada = ', k/ns)
```
