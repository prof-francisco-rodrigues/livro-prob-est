```python
# Pág. 249
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
n = 10000 # número de amostras
a = 1; b = 2 # valores de X devem ser > 0
# gera n valores de uma uniforme
X = np.random.uniform(a, b, n)
E = (a+b)/2 # valor esperado
vp = [] # armazena a probabilidade
ve = [] # armazena E[X]/delta
vd = [] # armazena os valores de delta
for delta in np.arange(a,b, 0.01):
    p = len(X[X >= delta])/len(X)
    vp.append(p) # armazena P(X>= delta)
    ve.append(E/delta) # armazena E[X]/delta
    vd.append(delta) # armazena delta
# mostra os resultados
plt.figure(figsize=(8,6))
plt.plot(vd, vp,linewidth=2, color='black', 
         linestyle = '--', label=r'$P(X \geq \delta)$')
plt.plot(vd, ve,linewidth=2, color='gray', 
         label=r'$E[X]/\delta$')
plt.xlabel(r'$\delta$' , fontsize = 20)
plt.legend(fontsize = 15)
plt.show(True)
```

```python
# Pág. 251
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(2011)
# gera as observações
mu = 10
sigma = 2
n = 10000
X = np.random.normal(mu, sigma, n)
E = np.mean(X)
vp = [] # armazena a probabilidade
ve = [] # armazena sigma2/delta2
vd = [] # armazena os valores de delta
for delta in np.arange(np.min(X),np.max(X), 0.05):
    # armazena P(|X-mu|)>= delta
    p = len(X[np.abs(X-mu) >= delta])/len(X)
    vp.append(p)
    # armazena sigma2/delta2
    ve.append(sigma**2/delta**2)
    # armazena delta
    vd.append(delta)
# mostra os resultados
plt.figure(figsize=(8,6))
plt.plot(vd, vp,linewidth=2, color='black', 
         linestyle = '--', label=r'$P(|X-\mu| > \delta)$')
plt.plot(vd, ve,linewidth=2, color='gray', 
         label=r'$\sigma^2/\delta^2$')
plt.xlabel(r'$\delta$' , fontsize = 20)
plt.legend(fontsize = 15)
plt.show(True)
```

```python
# Pág. 253
import matplotlib.pyplot as plt
import numpy as np

# função convexa
def g(x):
    g = x**2
    return g

N = 10000 # tamanho da amostra
vEg = [] # armazena E[g(X)]
vgE = [] # armazena g(E[X])
vn = range(10,N)
for n in vn:
    x = np.random.uniform(1,2,n)
    vEg.append(np.mean(g(x)))
    vgE.append(g(np.mean(x)))
# mostra os resultados
plt.figure(figsize=(8,6))
plt.plot(vn, vEg,linewidth=2, color='black', 
         linestyle = '-', label='E[g(X)]')
plt.plot(vn, vgE,linewidth=2, color='gray', label='g(E[X])')
plt.xlabel('Tamanho da amostra' , fontsize = 20)
plt.legend(fontsize = 15)
plt.show(True) 
```
