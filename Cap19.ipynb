```python
# Pág. 541
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(42)
rate = 10 # taxa
n = 30 # número de eventos
# armazena os tempos de ocorrência dos eventos
T = np.zeros(n) 
tn = 0 # tempo até o evento n
for i in range(n):# gera n eventos
    # sorteia o comprimento do intervalo
    ti = np.random.exponential(1/rate)
    tn = tn + ti
    T[i]=tn
plt.figure(figsize=(10,2))
plt.title("Tempo de cada evento", fontsize = 20)
for t in T:
    plt.axvline(x=t, linewidth = 2)
plt.xlabel('t', fontsize = 20) 
plt.xticks(fontsize=15)
plt.show(True)
```

```python
# Pág. 542
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(1974) # Define a semente

# Taxas dos três processos de Poisson 
rates  = [2, 5, 10] 
# Estilos de linha para diferenciar cada trajetória
styles = ['dotted', 'dashed', 'solid']
# Tempo máximo da simulação
tmax   = 10.0
plt.figure(figsize=(8,5))
# Para cada taxa, simulamos uma trajetória de N(t)
for rate, style in zip(rates, styles):
    # Iniciamos em t = 0 com N(0) = 0
    T = [0.0]   # tempos
    N = [0]     # número de eventos
    tn = 0.0 # tempo do último evento; 
    ne = 0 # número de eventos ocorridos até tn
    while True:
        tn = tn + np.random.exponential(1/rate)
        # Se o próximo evento ultrapassa o horizonte tmax, 
        # encerra a simulação desta trajetória
        if tn > tmax:
            break
        ne = ne + 1
        T.append(tn)
        N.append(ne)
    plt.step(T, N, where='post', linestyle=style, 
             label=rf'$\lambda={rate}$')

# Rótulos dos eixos e legenda
plt.xlim(0, tmax)
plt.xlabel('t', fontsize=15)
plt.ylabel('N(t)', fontsize=15)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Pág. 544
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(17)

plt.figure(figsize=(6,4))
lbd = 0.5 # taxa de nascimento
mu = 0.2 # taxa de morte
t = 100 # tempo atual
tf = 0 # tempo máximo
deltat = 0 # próximo intervalo entre eventos
vt = [] # armazena os tempos dos processos
N = 0 # população inicial
vN = [] # armazena a população no tempo t
while (tf + deltat) < t:
    deltat = np.random.exponential(1/(lbd+mu))
    tf = tf + deltat
    vt.append(tf)
    # ocorre um nascimento
    if (np.random.uniform() < lbd/(lbd+mu)):
        N = N + 1
    else:
    # ocorre uma morte
        N = max(0, N-1)
    vN.append(N)
# mostra os resultados
plt.step(vt, vN, linestyle = 'solid', color = 'gray')
plt.xlabel('t', fontsize=15)
plt.ylabel('N(t)', fontsize=15)
plt.show(True)
```
